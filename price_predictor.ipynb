# Train - Test  Splitting

```python
import numpy as np
```

```python
# for learning
def train_test_split(data,test_ratio):
    np.random.seed(42)
    shuffle = np.random.permutation(len(data))
    length_test_set = int(len(data)*test_ratio)
    test_indices = shuffle[:length_test_set]
    train_indices = shuffle[length_test_set:]
    return data.iloc[test_indices],data.iloc[train_indices]
    
```
